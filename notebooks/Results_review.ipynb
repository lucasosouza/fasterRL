# Experiment Report

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
%matplotlib inline

# statistics
from scipy import stats
from itertools import product, combinations
```

```python
experiments = [f for f in os.listdir("../logs/") if f[-4:] == 'json']
results = {}
for exp in experiments:
    exp_name = exp[:-5]
    try:
        with open("../results/" + exp, 'r') as f:
            results[exp_name] = json.load(f)
    except:
        pass
```

```python
def plot_variable(var):
    plt.figure(figsize=(13,5))
    plt.title("Distribution between no sharing and sharing agents")
    # plt.xticks(np.arange(0,300000,50000))
    plt.grid()
    for exp, vals in results.items():
        sns.kdeplot(vals[var], label=exp)
        
plot_variable("episodes_count")
```

```python
def calc_stats(var):
    # get specific variable to compare
    all_values = [(exp, values[var]) for exp, values in results.items()]
    samples = list(combinations(all_values, 2))
    for sample in samples:
        print(sample[0][0], sample[1][0])
        print(stats.ks_2samp(sample[0][1], sample[1][1]))  
        print("\n")

calc_stats("episodes_count")
```
